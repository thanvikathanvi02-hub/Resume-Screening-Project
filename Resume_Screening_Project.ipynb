```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
resumes = [
    "Python Machine Learning Data Science",
    "HTML CSS JavaScript Web Development",
    "Java Spring Boot Backend Development",
    "Figma UI UX Design",
    "Python AI Deep Learning",
    "React Frontend Development"
]

labels = [
    "Data Scientist",
    "Web Developer",
    "Backend Developer",
    "UI Designer",
    "Data Scientist",
    "Web Developer"
]
```

```python
cv = CountVectorizer()
X = cv.fit_transform(resumes)

X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.3, random_state=42)

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
cats = ["Data Scientist", "Web Developer", "Backend Developer", "UI Designer"]
count = [2, 2, 1, 1]

plt.bar(cats, count)
plt.title("Resume Category Distribution")
plt.xticks(rotation=30)
plt.show()
```

```python
test_text = ["Python Machine Learning"]
vec = cv.transform(test_text)

pred = model.predict(vec)[0]

roles = model.classes_

plt.bar(roles, [1 if r==pred else 0 for r in roles])
plt.title("Prediction Output Visualization")
plt.show()

print("Predicted Role:", pred)
```
